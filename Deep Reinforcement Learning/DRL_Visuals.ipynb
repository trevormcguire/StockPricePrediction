```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sys
```

```python
PATH = "PriceData"
MODEL_DIR = "FatCat"
TICKER = "AAPL"
epochs = 10
```

```python
memory = {}
for e in range(epochs + 1):
    e_name = "{}/FatCat_DRL_{}_mem_e{}.csv".format(MODEL_DIR, TICKER, e)
    mem_df = pd.read_csv(e_name).drop(columns=["Unnamed: 0"])
    memory[e] = mem_df
```

```python
for epoch, df in memory.items():
    buys, sells, holds = df[df.action == 1].index.values, df[df.action == 2].index.values, df[df.action == 0].index.values
    price_data = df.curr_price.values
    plt.figure(figsize=(15,10))
    plt.scatter(holds, price_data[holds], c="black", label="No Action")
    plt.scatter(buys, price_data[buys], c="green", label="Buy Signal")
    plt.scatter(sells, price_data[sells], c="red", label="Sell Signal")
    plt.legend()
    plt.title(f"{TICKER}: Game ({epoch})")
    plt.show()
```
